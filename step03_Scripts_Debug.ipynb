## 测试ImageFileReader()
以下是使用SimpleITK库中的`sitk.ImageFileReader()`的示例代码：

```python
import SimpleITK as sitk

# 创建图像文件读取器对象
reader = sitk.ImageFileReader()

# 设置要读取的图像文件路径
file_path = "path/to/your/image.nii.gz"
reader.SetFileName(file_path)

try:
    # 执行图像读取操作
    image = reader.Execute()

    # 可以对读取的图像进行后续处理或分析

    size = image.GetSize()

```

在上述示例中，首先创建一个`sitk.ImageFileReader()`对象，然后通过`SetFileName()`方法设置要读取的图像文件路径。接下来，使用`Execute()`方法执行图像读取操作，将读取的图像存储在`image`变量中。

请注意，示例中的`"path/to/your/image.nii.gz"`应替换为您实际的图像文件路径。另外，SimpleITK库支持多种图像格式，您可以根据实际情况选择正确的文件扩展名（例如`.nii.gz`、`.dcm`等）以匹配您的图像文件格式。

## 测试ImageSeriesReader()
以下是使用SimpleITK库中的`sitk.ImageSeriesReader()`的示例代码：

```python
import SimpleITK as sitk

# 创建图像序列读取器对象
reader = sitk.ImageSeriesReader()

# 设置要读取的图像系列的文件名格式
series_file_names = sitk.ImageSeriesReader.GetGDCMSeriesFileNames("path/to/your/series_folder")

# 将文件名列表传递给图像序列读取器
reader.SetFileNames(series_file_names)

try:
    # 执行图像序列读取操作
    image_series = reader.Execute()

    # 可以对读取的图像序列进行后续处理或分析
    # 例如，获取图像序列的大小和像素类型等信息
    size = image_series.GetSize()
    pixel_type = image_series.GetPixelIDTypeAsString()
    num_slices = image_series.GetSize()[2]

    # 打印图像序列信息
    print("Image series size:", size)
    print("Pixel type:", pixel_type)
    print("Number of slices:", num_slices)

except Exception as e:
    # 处理读取图像序列时的异常
    print("Error:", str(e))
```

在上述示例中，首先创建一个`sitk.ImageSeriesReader()`对象，然后使用`sitk.ImageSeriesReader.GetGDCMSeriesFileNames()`静态方法获取图像系列的文件名列表。该方法接受一个文件夹路径作为参数，根据所提供的文件名格式，自动识别并获取符合条件的图像文件名列表。

然后，将文件名列表传递给图像序列读取器的`SetFileNames()`方法，设置要读取的图像系列的文件名。

接下来，使用`Execute()`方法执行图像序列读取操作，将读取的图像序列存储在`image_series`变量中。

您可以根据需要对读取的图像序列进行后续处理或分析，例如获取图像序列的大小、像素类型以及切片数量等信息。

请注意，示例中的`"path/to/your/series_folder"`应替换为您实际的图像系列所在的文件夹路径。另外，`sitk.ImageSeriesReader`适用于读取DICOM格式的图像系列。如果您使用其他图像格式，请根据实际情况选择适当的图像读取器。

````python
import SimpleITK as sitk

# 读取dicom文件名
path_dicom = './Examples/abdominalDataSet/DICOM_anon'  # 存放dicom文件
reader_dicom = sitk.ImageSeriesReader()
name_dicom = reader_dicom.GetGDCMSeriesFileNames(path_dicom)
print('dicom',name_dicom)

# 读取nii文件名
path_nii = 'Examples/INPUT/'  # 存放nii文件
reader_nii = sitk.ImageSeriesReader()
name_nii = reader_dicom.GetGDCMSeriesFileNames(path_nii)
print('nii',name_nii)  # empty
````

````python
import SimpleITK as sitk

# 创建图像序列读取器对象
reader = sitk.ImageSeriesReader()

# 设置要读取的图像系列的文件名格式
# file_path = './Examples/INPUT/'
file_path = './Examples/abdominalDataSet/DICOM_anon'
series_file_names = sitk.ImageSeriesReader.GetGDCMSeriesFileNames(file_path)

# 将文件名列表传递给图像序列读取器
reader.SetFileNames(series_file_names)

try:
    # 执行图像序列读取操作
    image_series = reader.Execute()

    # 可以对读取的图像序列进行后续处理或分析
    # 例如，获取图像序列的大小和像素类型等信息
    size = image_series.GetSize()
    pixel_type = image_series.GetPixelIDTypeAsString()
    num_slices = image_series.GetSize()[2]

    # 打印图像序列信息
    print("Image series size:", size)
    print("Pixel type:", pixel_type)
    print("Number of slices:", num_slices)

except Exception as e:
    # 处理读取图像序列时的异常
    print("Error:", str(e))
````

## 结论
原因为传入的两个参数未以`/`结尾,即`./Examples/INPUT/`
1. ./Examples/INPUT/目录下文件后缀为`nii.gz`,为两个单独文件,由ImageFileReader()处理,而ImageSeriesReader()需要处理目录下的系列文件,如后缀为`dcm`的dicom文件,无法读取后缀为`nii`文件
2. 在`lung_extraction`中,执行`lung_extraction.py`文件,其中使用到CTLungSeg.utils._read_image()方法,代码块如下
    ```python
   def read_image(path):
    """读取path下Image Series/File"""
    if os.path.exists(path):  # 如果路径存在
        if os.path.isfile(path):  # 路径下是文件,调用ImageFile读取器
            reader = _read_image(path)
        else:  # 路径下是文件夹,调用ImageSeries读取器
            reader = _read_image_series(path)

        image = reader.Execute()  # 执行图像读取操作
    else:  # 路径不存在抛出异常
        raise FileNotFoundError(f"Cannot find {path}")

    return image
    ```
   其通过判断提供**路径**为文件(单个文件)或目录(文件组),进而决定使用`ImageFileReader()`或`ImageSeriesReader()`,而当文件传入地址不正确时,无法对地址内容进行判断,进而无法决定调用`ImageFileReader()`或`ImageSeriesReader()`,从而抛出异常.
